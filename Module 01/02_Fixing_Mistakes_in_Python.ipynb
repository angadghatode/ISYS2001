# Breaking Things on Purpose

Good news: you're about to write some code that doesn't work. Also good news — that's the entire point of this notebook.

Every programmer, no matter how many years they've been at it, spends a chunk of their day staring at red error messages. The skill that actually matters isn't avoiding errors, it's reading them calmly and knowing where to look. Python is unusually generous about telling you what went wrong — you just have to get comfortable listening to it.

In this notebook, you'll deliberately run broken code, read Python's error message, and fix it. Every mistake below is one that almost everyone makes in their first week:

- Getting the capitalisation of a command wrong
- Losing a bracket or a quotation mark
- Indenting a line that shouldn't be indented
- A plain old typo

> 💡 **Advanced note:** This process of finding and fixing mistakes is called **debugging**. You'll get noticeably faster at it as the semester goes on.

## Section 4: Python Cares About Capitals

Python treats `print`, `Print`, and `PRINT` as three completely different things — and only one of them actually works. Run the code below to see the error, then fix it.


```python
PRINT("All caps")
```

> 💡 Reading tip: you won't understand every line of a Python error message, and that's normal. Skim for the parts in plain English and the line number it points to.
