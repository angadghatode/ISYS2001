## Part 2: Your First AI Prompt

Ask an AI assistant (Gemini in Colab, ChatGPT, or GitHub Copilot) to write a program that:
- Asks the user for their name
- Asks for their monthly income
- Prints a personalised welcome message

Write your prompt as a comment, then paste in whatever code the AI gives you back.


```python
# Write your AI prompt here as a comment:
"""
My AI Prompt:
Write a simple Python script for a command-line budget tracker. It needs to do exactly these three things:

Prompt the user to input their name.
Prompt the user to input their monthly income (make sure it handles numbers/decimals).
Print a personalized welcome message that includes their name and confirms their monthly income.

"""

# Paste the AI-generated code below:
def main():
    # 1. Prompt the user for their name
    name = input("Please enter your name: ").strip()

    # 2. Prompt the user for their monthly income and handle decimals
    while True:
        try:
            income_input = input("Please enter your monthly income: $")
            income = float(income_input)

            if income < 0:
                print("Income cannot be negative. Please try again.")
                continue

            break  # Exit the loop once a valid number is entered
        except ValueError:
            print("Invalid input. Please enter a valid number (e.g., 4000 or 4500.50).")

    # 3. Print the personalized welcome message
    print(f"\nWelcome to your Command-Line Budget Tracker, {name}!")
    print(f"Your monthly income has been successfully set to ${income:,.2f}.")

if __name__ == "__main__":
    main()
```

## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income
- Takes up to 3 expenses
- Calculates what's left over
- Warns the user if the budget goes negative

> 💡 **Pro tip:** The more specific your prompt, the less clean-up you'll need to do afterwards. "Write a budget tracker" gets you something generic; describing the exact inputs, outputs, and edge cases gets you something closer to what you actually need.


```python
# Write your detailed prompt here:
"""
My Budget Tracker Prompt:
Write a Python script for a simple command-line budget tracker. Please follow these exact requirements:
1. Inputs:
Prompt the user for their total monthly income (ensure it accepts decimals).
Prompt the user to enter up to 3 expenses. For each expense, ask for the name of the expense and the amount.
Allow the user to stop entering expenses early if they have fewer than 3 (e.g., by typing 'done').
2. Calculations:
Calculate the total of the entered expenses.
Subtract the total expenses from the monthly income to calculate the remaining balance.
3. Outputs & Warnings:
Print a clean summary showing the total expenses and the leftover amount.
Edge Case: If the remaining balance goes below zero, print a clear warning message (e.g., "Warning: You are over budget!").
Error Handling: Ensure the program handles invalid inputs (like someone typing letters instead of numbers for their income) without crashing.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
def main():
    print("--- Command-Line Budget Tracker ---\n")

    # 1. Prompt for monthly income with error handling
    while True:
        try:
            income_input = input("Enter your total monthly income: $")
            income = float(income_input)
            if income < 0:
                print("Income cannot be negative. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid number (e.g., 3000 or 4500.50).")

    total_expenses = 0.0

    # 2. Prompt for up to 3 expenses
    print("\nEnter up to 3 expenses. (Type 'done' for the expense name to stop early)")

    for i in range(1, 4):
        expense_name = input(f"\nExpense {i} name: ").strip()

        # Check for early exit
        if expense_name.lower() == 'done':
            break

        # Prompt for expense amount with error handling
        while True:
            try:
                amount_input = input(f"Enter amount for '{expense_name}': $")
                amount = float(amount_input)
                if amount < 0:
                    print("Expense amount cannot be negative. Please try again.")
                    continue

                total_expenses += amount
                break
            except ValueError:
                print("Invalid input. Please enter a numerical value.")

    # 3. Calculations
    remaining_balance = income - total_expenses

    # 4. Outputs & Warnings
    print("\n" + "="*30)
    print("       BUDGET SUMMARY")
    print("="*30)
    print(f"Total Income:   ${income:,.2f}")
    print(f"Total Expenses: ${total_expenses:,.2f}")
    print(f"Leftover Funds: ${remaining_balance:,.2f}")

    # Check for over-budget edge case
    if remaining_balance < 0:
        print("\n⚠️ WARNING: You are over budget!")
    print("="*30)

if __name__ == "__main__":
    main()
```

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)

- 📝 A detailed budget summary with advice


```python
# Document your enhancement choice and the prompt you used
enhancement_choice = "I chose to add: Percentage of income spent"
enhancement_prompt = """
Can you add a feature to this code for the code to be able to show percentage
of income spent of total income at the end within the summary.
"""

print(f"🚀 Enhancement: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
def main():
    print("--- Command-Line Budget Tracker ---\n")

    # 1. Prompt for monthly income with error handling
    while True:
        try:
            income_input = input("Enter your total monthly income: $")
            income = float(income_input)
            if income < 0:
                print("Income cannot be negative. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid number (e.g., 3000 or 4500.50).")

    total_expenses = 0.0

    # 2. Prompt for up to 3+  expenses
    print("\nEnter up to 3+ (4) expenses. (Type 'done' for the expense name to stop early)")

    for i in range(1, 5):
        expense_name = input(f"\nExpense {i} name: ").strip()

        # Check for early exit
        if expense_name.lower() == 'done':
            break

        # Prompt for expense amount with error handling
        while True:
            try:
                amount_input = input(f"Enter amount for '{expense_name}': $")
                amount = float(amount_input)
                if amount < 0:
                    print("Expense amount cannot be negative. Please try again.")
                    continue

                total_expenses += amount
                break
            except ValueError:
                print("Invalid input. Please enter a numerical value.")

    # 3. Calculations
    remaining_balance = income - total_expenses

    # Calculate percentage spent (preventing division by zero)
    if income > 0:
        percentage_spent = (total_expenses / income) * 100
    else:
        percentage_spent = 0.0

    # 4. Outputs & Warnings
    print("\n" + "="*30)
    print("       BUDGET SUMMARY")
    print("="*30)
    print(f"Total Income:     ${income:,.2f}")
    print(f"Total Expenses:   ${total_expenses:,.2f}")
    print(f"Leftover Funds:   ${remaining_balance:,.2f}")
    print(f"Percentage Spent: {percentage_spent:.1f}%")

    # Check for over-budget edge case
    if remaining_balance < 0:
        print("\n⚠️ WARNING: You are over budget!")
    print("="*30)

if __name__ == "__main__":
    main()
```
